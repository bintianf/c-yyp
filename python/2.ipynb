```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.interpolate

file=open('../../map.txt')
data = file.readlines()
b = []
u = []
v = []
for num in data:
    b.append(int(num.split(',')[0]))
    u.append(int(num.split(',')[1]))
    v.append(int(num.split(',')[2]))
x = np.arange(0,100,1)
y = np.arange(0,100,1)
U = [[0]*100 for i in range(100) ]
V = [[0]*100 for i in range(100) ]
for i in range(100):
    for j in range(100):
        pos = i + j*100
        U[i][j] = u[pos]
        V[i][j] = v[pos]
# xi, yi = np.linspace(x.min(), x.max(), 10), np.linspace(y.min(), y.max(), 10)
# xi, yi = np.meshgrid(x, y)
# rbf = scipy.interpolate.Rbf(x, y, u, function='linear')
# zi = rbf(xi, yi)
# plt.contour(xi, yi, zi, colors='black')
X,Y=np.meshgrid(x,y)

px = range(0, 100, 10)
py = range(0, 100, 10)
plt.xticks(px)
plt.yticks(py)
plt.contour(X,Y,U)  #生成等值线图
plt.contourf(X,Y,U)  #生成等值线图
plt.savefig("1.png", dpi = 1400)
plt.show()
```
